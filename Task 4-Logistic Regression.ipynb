### Import Required Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, classification_report, roc_auc_score, roc_curve, precision_score, recall_score
```

### Load and Explore the Dataset

```python
# Load the dataset
df=pd.read_csv("Breast Cancer.csv")
```

```python
df.drop(columns=['Unnamed: 32'], inplace=True)
```

```python
# Convert categorical 'diagnosis' column to numeric (M = 1, B = 0)
df['diagnosis'] = df['diagnosis'].map({'M': 1, 'B': 0})
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(x=df['diagnosis'], palette='coolwarm')
plt.title("Class Distribution (Malignant vs Benign)")
plt.xlabel("Diagnosis")
plt.ylabel("Count")
plt.show()
```

```python
# Encode the diagnosis column (Malignant -> 1, Benign -> 0)
label_encoder = LabelEncoder()
df['diagnosis'] = label_encoder.fit_transform(df['diagnosis'])
```

```python
# Drop ID column
if 'id' in df.columns:
    df.drop(columns=['id'], inplace=True)
```

### Feature Selection

```python
# Select specific features
selected_features = ['radius_mean', 'texture_mean']
X_selected = df[selected_features].values  # Features
```

### Data Preprocessing

```python
# Encode the target variable
label_encoder = LabelEncoder()
y_selected = label_encoder.fit_transform(df['diagnosis'].values)  # Target variable
```

```python
# Train/test split
X_train, X_test, y_train, y_test = train_test_split(X_selected, y_selected, test_size=0.2, random_state=42)
```

```python
# Standardize features
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# Create and fit the model
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
# Make predictions
y_pred = model.predict(X_test)
y_pred_prob = model.predict_proba(X_test)[:, 1]
```

```python
# Evaluate the model
conf_matrix = confusion_matrix(y_test, y_pred)
class_report = classification_report(y_test, y_pred)
roc_auc = roc_auc_score(y_test, y_pred_prob)

print(f'Confusion Matrix:\n{conf_matrix}')
print(f'Classification Report:\n{class_report}')
print(f'ROC-AUC: {roc_auc}')
```

```python
# Plotting the ROC Curve
fpr, tpr, thresholds = roc_curve(y_test, y_pred_prob)
plt.plot(fpr, tpr, label='Logistic Regression (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], 'k--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc='lower right')
plt.show()
```

```python
# Probability Histogram
plt.subplot(1, 3, 3)
plt.hist(y_pred_prob[y_test == 1], bins=20, alpha=0.6, label='Benign (1)', color='green')
plt.hist(y_pred_prob[y_test == 0], bins=20, alpha=0.6, label='Malignant (0)', color='red')
plt.xlabel('Predicted Probability')
plt.ylabel('Frequency')
plt.title('Prediction Probabilities')
plt.legend()
```

### Threshold Tuning

```python
# Tuning the threshold
threshold = 0.5  # Default threshold
y_pred_tuned = (y_pred_prob >= threshold).astype(int)

# Evaluate with the new threshold
conf_matrix_tuned = confusion_matrix(y_test, y_pred_tuned)
print(f'Tuned Confusion Matrix:\n{conf_matrix_tuned}')
```
